```python
import numpy as np
import pandas as pd
from pathlib import Path
import seaborn as sns
```

#Piloto

```python
# Get participant files name
participant_files = []
for p in Path("./").glob("*.csv"):
    participant_files.append(f"{p.name}")

# Create empty DataFrame
data_piloto = pd.DataFrame()

# Compute data for each participant
for participant_file in participant_files:

  data = pd.read_csv(participant_file)
  data = data[
      [
          "participant",
          "mouse.clicked_name",
       ]
  ]

  data = data.dropna()

  par_df = pd.DataFrame()

  par = {
    "Participant": data.participant.iloc[-1],
    "Izquieda_Pr": data["mouse.clicked_name"].value_counts()["['image']"],
    "Derecha_Pr": data["mouse.clicked_name"].value_counts()["['image_2']"],
    }


  new_df = pd.DataFrame(data=par, index=[5])
  par_df = pd.concat([par_df, new_df], ignore_index=True)

  # Append to final Dataframe
  data_piloto = pd.concat([data_piloto, par_df], ignore_index=True)

data_piloto
```

```python
# Get participant files name
participant_files = []
for p in Path("./").glob("*.csv"):
    participant_files.append(f"{p.name}")

# Create empty DataFrame
data_metaforas = pd.DataFrame()

# Compute data for each participant
for participant_file in participant_files:

  data = pd.read_csv(participant_file)
  data = data[
      [
          "participant",
          "mouse_6.clicked_name",
       ]
  ]

  data = data.dropna()

  par_df = pd.DataFrame()

  par = {
    "Participant": data.participant.iloc[-1],
    "Derecha_T": data["mouse_6.clicked_name"].value_counts()["['derecha']"],
    "Izquierda_T": data["mouse_6.clicked_name"].value_counts()["['izquierda']"],
    }


  new_df = pd.DataFrame(data=par, index=[5])
  par_df = pd.concat([par_df, new_df], ignore_index=True)

  # Append to final Dataframe
  data_metaforas = pd.concat([data_metaforas, par_df], ignore_index=True)

data_metaforas
```
