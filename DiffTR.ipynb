```python
import pandas as pd
from pathlib import Path

# Get participant files name
participant_files = [p.name for p in Path("./").glob("*.csv")]

# Create empty DataFrame
psm = pd.DataFrame()

# Compute data for each participant
for participant_file in participant_files:

  data = pd.read_csv(participant_file)
  data = data[
      [
          "participant",
          "key_resp_17.corr",
          "ADQUISICIÓN_.stopped",
          "key_resp_17.started",
       ]
  ]
  data = data.dropna()

  par_df = pd.DataFrame()

  rt = (data["ADQUISICIÓN_.stopped"] - data["key_resp_17.started"]).mean()
  corr_count = data["key_resp_17.corr"].sum()

  par = {
    "Participant": data.participant.iloc[-1],
    "RT": rt,
    "Correct": corr_count,
    }

  new_df = pd.DataFrame(data=par, index=[5])
  par_df = pd.concat([par_df, new_df], ignore_index=True)

    # Append to final DataFrame
  psm = pd.concat([psm, par_df], ignore_index=True)

psm
```
