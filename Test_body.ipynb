```python
import pandas as pd
from pathlib import Path

# Get participant files name
participant_files = [p.name for p in Path("./").glob("*.csv")]

# Create empty DataFrame
data_flanker = pd.DataFrame()

# Compute data for each participant
for participant_file in participant_files:
    data = pd.read_csv(participant_file)
    data = data[["participant", "Nivel", "key_resp.corr", "key_resp.rt"]]
    data = data.dropna()

    par_df = pd.DataFrame()

    # Data by Condition "Nivel" (independent variable)
    for Nivel in data.Nivel.unique():
        nivel_df = data.query("Nivel == @Nivel")
        rt = nivel_df["key_resp.rt"].mean()
        corr_count = nivel_df["key_resp.corr"].sum()

        par = {
            "Participant": data.participant.iloc[0],
            "Nivel": Nivel,
            "RT": rt,
            "Correct": corr_count,
        }

        new_df = pd.DataFrame(par, index=[0])
        par_df = pd.concat([par_df, new_df], ignore_index=True)

    # Append to final DataFrame
    data_flanker = pd.concat([data_flanker, par_df], ignore_index=True)

data_flanker
```
